**Cleaning bayern_summary.csv and ameding dates to leave only month Sept for each year**


```python
import pandas as pd
df = pd.read_csv("bayern_summary_v3.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter rows where the month is September and Bayern Munich is the home or away team
filtered_data = df[(df['Date'].dt.month == 9) & ((df['HomeTeam'] == 'Bayern Munich') | (df['AwayTeam'] == 'Bayern Munich'))]

filtered_data 
```

```python
# Extract the year from the 'Date' column using .loc to avoid the SettingWithCopyWarning
filtered_data.loc[:, 'Year'] = filtered_data['Date'].dt.year

# Group by year and count occurrences in FTHG and FTAG
bayern_summary = filtered_data.groupby('Year').agg(
    TotalGoalsAsHomeTeam=('FTHG', 'sum'),
    TotalGoalsAsAwayTeam=('FTAG', 'sum')
).reset_index()

# DataFrame
bayern_summary
```

```python
bayern_summary['sum_column'] = bayern_summary['TotalGoalsAsHomeTeam'] + bayern_summary['TotalGoalsAsAwayTeam']
bayern_summary
```

```python
bayern_summary.dtypes
```

```python
columns_export = ["Year", "sum_column" ]
bayern_summary_v3 = bayern_summary[columns_export]
bayern_summary_v3.to_csv('bayern_summary_v3.csv',index=False)
```

```python
bayern_summary.to_csv('bayern_summary.csv', index=False)
```
