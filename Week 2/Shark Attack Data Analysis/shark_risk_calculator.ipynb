```python
# This is a calculator to estimate risk based on inputs of country, time of year, activity and age 

country = input("Which country are you visiting? [Australia, South Africa, or USA] ").lower()

if country in ["australia", "south africa", "usa"]:
    while True:
        month_of_year = input("Which month of the year are you planning to visit? ").lower()

        # Dictionary to store messages for each month
        month_messages = {
            "january": "I hear it's lovely at that time of year.",
            "february": "I hear it's lovely at that time of year.",
            "march": "I hear it's lovely at that time of year.",
            "april": "I hear it's lovely at that time of year.",
            "may": "I hear it's lovely at that time of year.",
            "june": "I hear it's lovely at that time of year.",
            "july": "I hear it's lovely at that time of year.",
            "august": "I hear it's lovely at that time of year.",
            "september": "I hear it's lovely at that time of year.",
            "october": "I hear it's lovely at that time of year.",
            "november": "I hear it's lovely at that time of year.",
            "december": "I hear it's lovely at that time of year."
        }

        # Check if the entered month is valid
        if month_of_year in month_messages:
            print(f"You're planning to visit {country.capitalize()} in {month_of_year.capitalize()}. {month_messages[month_of_year]}")

            activity = input("Which watersport are you planning to do there? [boarding, fishing, diving, or something else] ").lower()

            # Initialize risk calculator
            risk_calculator = 0

            # Update risk calculator based on activity
            if activity == "diving":
                risk_calculator += 4
            elif activity == "fishing":
                risk_calculator += 3
            elif activity == "boarding":
                risk_calculator += 2
            else:
                risk_calculator += 1

            age = input("One last question: how many years young are you? [<20, 21-30, 31-40, 41-50, 51-60, 61+] ")

            # Update risk calculator based on age
            if age == "<20":
                risk_calculator += 1
            elif age == "21-30":
                risk_calculator += 1
            elif age == "31-40":
                risk_calculator += 2
            elif age == "41-50":
                risk_calculator += 2
            elif age == "51-60":
                risk_calculator += 3
            elif age == "61+":
                risk_calculator += 3
            else:
                risk_calculator += 1

            # Update risk calculator based on summer month
            if country == "australia" and month_of_year in ["december", "january", "february"]:
                risk_calculator += 5
            elif country == "usa" and month_of_year in ["june", "july", "august"]:
                risk_calculator += 1
            elif country == "south africa" and month_of_year in ["december", "january", "february"]:
                risk_calculator += 3

            print(f"Risk Calculator: {risk_calculator} points")

            # Categorize risk level
            if risk_calculator <= 5:
                risk_level = "Low Risk; sounds like you might need to spice your holiday up a bit"
            elif 6 <= risk_calculator <= 9:
                risk_level = "Medium Risk; you should be fine, maybe bring a spear to be on the safe side"
            else:
                risk_level = "High Risk; here's a voucher code for our will writing partner: XCR5678 "

            print(f"Risk Level: {risk_level}")

            break  # Exit the loop when a valid month, activity, and age are entered
        else:
            print("That doesn't look like a month. Please enter a valid month.")
else:
    print("You're not planning to visit Australia, South Africa, or the USA.")
```
