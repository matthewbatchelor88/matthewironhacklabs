```python
import pandas as pd
```

```python
# Load the data

url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

df_1 = pd.read_csv(url_1)
df_2 = pd.read_csv(url_2)
df_3 = pd.read_csv(url_3)
```

```python
# Merge the three data sets vertically (i.e. just adding more instances)

pd.concat([df_1, df_2, df_3], axis=0)
```

```python
# Create new data frame of concatenated data frames 

new_df = pd.concat([df_1, df_2, df_3], axis=0)
```

```python
# Take steps to clean the data 

def make_headers_lowercase(df):
    # Make column headers lowercase
    df.columns = df.columns.str.lower()
    return df
```

```python
make_headers_lowercase(new_df)
```

```python
new_df.rename(columns= {'st': 'state'}, inplace=True)
print(new_df.columns)
```

```python
new_df['gender'] = new_df['gender'].replace("Femal", "F").replace("Male", "M").replace("female","F")
```

```python
new_df['gender'].describe()
```

```python
new_df['state']=new_df['state'].replace("AZ", "Arizona"). replace("Cali","California").replace("WA", "Washington").replace("NA", "Nevada")
```

```python
new_df['education']=new_df['education'].replace("Bachelors", "Bachelor")
```

```python
new_df['customer lifetime value'] = pd.to_numeric(new_df['customer lifetime value'].str.rstrip('%'), errors='coerce')
```

```python
new_df['vehicle class']=new_df['vehicle class'].replace("Sports Car", "Luxury").replace("Luxury SUV", "Luxury").replace("Luxury Car", "Luxury")
```

```python
new_df['number of open complaints'] = new_df['number of open complaints'].apply(lambda x: int(x.split("/")[1]) if isinstance(x, str) and "/" in x else x)
```

```python
# A null value in 'customer' implies no customer id; it is meaningful to group these customers by a 'no id' attribute

new_df['customer'] = new_df['customer'].fillna('no id')
```

```python
# As the group of customers with no 'state' value represent a majority of customers, it would be overly drastic to exclude them. As such, we apply 'state unknown'

new_df['state'] = new_df['state'].fillna('state unknown')
```

```python
new_df['gender'] = new_df['gender'].fillna('gender unknown')
```

```python
new_df['education'] = new_df['education'].fillna('High School or Below')
```

```python
# On customer lifetime value, it is reasonable to assume that a null value equals 0 in most cases

new_df['customer lifetime value'] = new_df['customer lifetime value'].fillna('0')
```

```python
# On income, it is reasonable to assume that a null value equals 0 in most cases

new_df['income'] = new_df['income'].fillna('0')
```

```python
# On monthly premium auto, it is reasonable to assume that a null value equals 0 in most cases

new_df['monthly premium auto'] = new_df['monthly premium auto'].fillna('0')
```

```python
# On number of open complaints, it is reasonable to assume that a null value equals 0 in most cases

new_df['number of open complaints'] = new_df['number of open complaints'].fillna('0')
```

```python
# A null value in 'policy type' implies that customer has no active policy; it is meaningful to group these customers by a 'no active policy' attribute

new_df['policy type'] = new_df['policy type'].fillna('no active policy')
```

```python
# A null value in 'vehicle class' implies vehicle is unknown

new_df['vehicle class'] = new_df['vehicle class'].fillna('vehicle unknown')
```

```python
# On total claim amount, it is reasonable to assume that a null value equals 0 in most cases

new_df['total claim amount'] = new_df['total claim amount'].fillna('0')
```

```python
new_df_no_duplicate_rows = new_df.drop_duplicates(subset=['customer'])
```

```python
new_df
```

```python
new_df_no_duplicate_rows
```

```python
# Removing duplicates in subset customer reduces row count by 2939, implying 2939 duplicate customers when combining data sets 
```

```python
# Simplify df name for readability 

new_df_v2 = new_df_no_duplicate_rows
```

```python
# Now need to resolve duplicate columns 

# Identify duplicate columns
duplicate_columns = new_df_v2.columns[new_df_v2.columns.duplicated()].tolist()
duplicate_columns
```

```python
df_no_duplicate_columns = new_df_v2.copy()

# Get a list of duplicate column names
duplicate_columns = new_df_v2.columns[new_df_v2.columns.duplicated(keep='first')]

# Iterate over duplicate columns and merge them, retaining the left instance where 'state unknown' is present
for col in duplicate_columns:
    # Combine values from duplicate columns, keeping values from 'state' where 'state unknown' is present
    df_no_duplicate_columns[col] = new_df_v2.filter(like=col).apply(lambda x: x.combine_first(x.dropna()))

# Drop the duplicate columns
df_no_duplicate_columns = df_no_duplicate_columns.loc[:, ~df_no_duplicate_columns.columns.duplicated()]
```
